## Görev 1: Pasta Grafikleriyle Veri Bileşimini Görselleştirme
***1.1 Katılımcıların Çalışmak İstediği En İyi 5 Veritabanının Pasta Grafiğini Oluşturma
Anket verilerinde, DatabaseWantToWorkWith sütununda katılımcıların çalışmak istediği veritabanları listelenmiştir. En çok istenen 5 veritabanını bir pasta grafiğinde görselleştirelim.***

```python
print(df['DatabaseWantToWorkWith'].value_counts())
print('----------------**********************----------******************')

# 1. Veriyi temizle ve ayır (Explode)
df_db = df['DatabaseWantToWorkWith'].str.split(';').explode()
db_counts = df_db.value_counts().head(5)
print(db_counts)



# 2. Pasta Grafiği
plt.figure(figsize=(10, 8))
plt.pie(db_counts, labels=db_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('En Çok Çalışılmak İstenen İlk 5 Veritabanı', fontsize=14)
plt.show()
```

***1.2 "DevType" sütunu, katılımcıların geliştirici türlerini listeler. En yaygın 5 geliştirici rolünü pasta grafiğinde göstererek dağılımı inceleyeceğiz.***

```python
print(df['DevType'].value_counts())
print('----------------**********************----------******************')

# 1. En popüler 5 rolü al
dev_counts = df['DevType'].value_counts().head(5)

# 2. Pasta Grafiği
plt.figure(figsize=(10, 8))
# explode ile en büyük dilimi (Full-stack) biraz dışarı çıkaralım (vurgu için)
explode = (0.1, 0, 0, 0, 0) 

plt.pie(dev_counts, labels=dev_counts.index, autopct='%1.1f%%', 
        startangle=90, colors=plt.cm.Pastel1.colors, explode=explode, shadow=True)

plt.title('En Yaygın 5 Geliştirici Rolü Dağılımı', fontsize=14)
plt.show()
```

***1.3 Katılımcıların profesyonel kullanım için kullandıkları işletim sistemlerine ilişkin bir pasta grafiği oluşturun***
OpSysProfesyonel kullanım sütunu, geliştiricilerin profesyonel olarak kullandıkları işletim sistemlerini göstermektedir. En çok kullanılan işletim sistemlerinin dağılımını bir pasta grafiğinde görselleştirelim.

```python
# 1. Veriyi temizle ve ayır (İşletim sistemleri de genelde noktalı virgülle ayrılır)
df_opsys = df['OpSysProfessional use'].str.split(';').explode()
opsys_counts = df_opsys.value_counts().head(5)

# 2. Pasta Grafiği
plt.figure(figsize=(10, 8))
plt.pie(opsys_counts, labels=opsys_counts.index, autopct='%1.1f%%', 
        startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'], shadow=True)

plt.title('Profesyonel Kullanım İçin En Çok Tercih Edilen 5 İşletim Sistemi', fontsize=14)
plt.show()
```

# Görev 2: Ek Görselleştirmeler ve Karşılaştırmalar
***2.1 Katılımcıların Çalıştığı En İyi 5 Programlama Dili için Pasta Grafiği
LanguageHaveWorkedWith sütunu, katılımcıların deneyim sahibi olduğu programlama dillerini içerir. En iyi 5 dilin bileşimini göstermek için bir pasta grafiği çizeceğiz.***

```python

# 1. Dilleri ayır ve her birini tekil olarak say (Explode)
df_languages = df['LanguageHaveWorkedWith'].str.split(';').explode()
lang_counts = df_languages.value_counts().head(5)

# 2. Pasta Grafiği Oluşturma
plt.figure(figsize=(10, 8))
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'] # Modern renk paleti

plt.pie(lang_counts, 
        labels=lang_counts.index, 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=colors, 
        wedgeprops={'edgecolor': 'black'}, shadow=True) # Dilimler arasına ince beyaz çizgi

plt.title('Katılımcıların Çalıştığı En Popüler 5 Programlama Dili', fontsize=14)
plt.show()
```

***2.2 Profesyonel Kullanımda En Çok Kullanılan İş Birliği Araçları için Pasta Grafiği
NEWCollabToolsHaveWorkedWith sütununu kullanarak, katılımcıların profesyonel çalışmalarında en çok kullandıkları iş birliği araçlarını belirleyip görselleştireceğiz.***

```python
# 1. Araçları ayır ve tek tek say
df_tools = df['NEWCollabToolsHaveWorkedWith'].str.split(';').explode()
tool_counts = df_tools.value_counts().head(5)

# 2. Pasta Grafiği Oluşturma
plt.figure(figsize=(10, 8))

# En popüler aracı (genelde VS Code) vurgulamak için hafifçe dışarı çıkaralım
explode = (0.05, 0, 0, 0, 0) 

plt.pie(tool_counts, 
        labels=tool_counts.index, 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=plt.cm.Set3.colors, 
        explode=explode, 
        shadow=True)

plt.title('Profesyonel Kullanımda En Çok Tercih Edilen 5 İş Birliği Aracı', fontsize=14)
plt.show()
```

## Görev 3: Kompozisyonu Analiz Etme ve Yorumlama
Bu görevde, anket verilerinin belirli yönlerini analiz etmek için ek pasta grafikler oluşturacaksınız. Her görevi tamamlamak ve bulguları yorumlamak için pandas ve matplotlib kullanın.

***3.1 Katılımcıların En Çok Beğendiği Programlama Dilleri Pasta Grafiği
LanguageAdmired sütunu, katılımcıların en çok beğendiği programlama dillerini listeler. En çok beğenilen 5 dili görselleştirmek için bir pasta grafik oluşturun.***

```python

# 1. Veriyi temizle ve ayır
# dropna() ekleyerek boş satırların hata yapmasını engelleyelim
df_admired = df['LanguageAdmired'].dropna().str.split(';').explode()
admired_counts = df_admired.value_counts().head(5)

# Eğer veri boş değilse çizim yap
if not admired_counts.empty:
    plt.figure(figsize=(10, 7))
    
    # Renk paletini daha güvenli bir yolla tanımlayalım
    colors = plt.get_cmap('Spectral')(range(0, 256, 50)) 

    # Çizim (Hata payını azaltmak için veriyi ve etiketleri net veriyoruz)
    plt.pie(x=admired_counts.values, 
            labels=admired_counts.index, 
            autopct='%1.1f%%', 
            startangle=140, 
            colors=colors, 
            wedgeprops={'edgecolor': 'white', 'linewidth': 2})
    
    plt.title('Katılımcıların En Çok Hayranlık Duyduğu 5 Dil', fontsize=14)
    plt.axis('equal') # Pastanın elips değil tam daire görünmesini sağlar
    plt.show()
else:
    print("Görselleştirilecek veri bulunamadı! Lütfen LanguageAdmired sütununu kontrol edin.")
```

***3.2 Yapay Zeka Geliştirmede Kullanılan Araçların Pasta Grafiği
"AIToolCurrently Using" sütununu kullanarak, geliştiricilerin şu anda yapay zeka geliştirmede kullandığı en iyi 5 aracı görselleştirmek için bir pasta grafiği oluşturun.***

```python
# 1. AI kullanım amaçlarını ayır ve say
df_ai = df['AIToolCurrently Using'].str.split(';').explode()
ai_counts = df_ai.value_counts().head(5)

# 2. Pasta Grafiği
plt.figure(figsize=(10, 7))
plt.pie(ai_counts, labels=ai_counts.index, autopct='%1.1f%%', 
        startangle=90, colors=plt.cm.Accent.colors)
plt.title('AI Araçlarının En Çok Kullanıldığı 5 Alan', fontsize=14)
plt.show()
```

EmbeddedWantToWorkWith sütununu kullanarak, katılımcıların çalışmak istediği en çok arzu edilen 5 gömülü sistem teknolojisini gösteren bir pasta grafiği oluşturun.***
