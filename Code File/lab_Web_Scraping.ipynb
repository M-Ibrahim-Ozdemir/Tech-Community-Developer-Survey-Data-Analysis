Kazımanız gereken veriler, programlama dilinin adı ve ortalama yıllık maaştır.
Kazımaya başlamadan önce URL'yi web tarayıcınızda açıp web sayfasının içeriğini incelemeniz iyi bir fikirdir.

Gerekli kütüphaneleri içe aktarın.

```python
from bs4 import BeautifulSoup # HTML'i anlamlandırmak için
import requests  # Web sayfasını indirmek için
import pandas as pd # Verileri tablo yapmak için
```

Aşağıdaki bağlantıdan web sayfasını indirin.

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
data = requests.get(url).text # Sayfanın metin içeriğini alıyoruz
```

Bir "soup" nesnesi oluşturun.

```python
soup = BeautifulSoup(data, "html.parser")
```

Dil adını ve yıllık ortalama maaşı kazıyın.

```python
# Önce boş bir liste oluşturalım
language_list = []

# Tablodaki tüm satırları bulalım
tabel = soup.find('table')

for row in tabel.find_all('tr'):
    cols = row.find_all('td') # Her satırdaki sütunları (td) alıyoruz
    
    # Tablo başlığını atlamak ve sadece verileri almak için kontrol
    language = cols[1].getText()  # Dil adı 2. sütunda
    salary = cols[3].getText()    # Maaş 4. sütunda
    
    # Verileri listemize ekliyoruz
    language_list.append([language, salary])
    print(f"{language} ----> {salary}")
```

Kazınan verileri "popüler-diller.csv" adlı bir dosyaya kaydedin.

```python
# Listeyi bir Pandas DataFrame'e çevirelim
df =  pd.DataFrame(language_list[1:], columns=["Language", "Avrage Annual Salary"])

# CSV olarak kaydet
df.to_csv("populer-language.csv", index=False)
print("\nVeriler 'popular-languages.csv' adıyla başarıyla kaydedildi!")
```
