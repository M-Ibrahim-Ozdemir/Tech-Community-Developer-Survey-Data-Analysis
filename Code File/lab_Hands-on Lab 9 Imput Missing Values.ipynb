# Adım 1: Gerekli Kütüphaneleri İçe Aktarın¶


```python
import pandas as pd
```

## Adım 2: Veri Setini Bir Veri Çerçevesine Yükleyin
Verileri Okuyun
Aşağıdaki fonksiyonlar veri setini tarayıcınıza indirecektir:

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```


## 3. Adım. Yinelenenleri Bulma ve Kaldırma
Görev 1: Veri kümesindeki yinelenen satırları belirleyin.

# 1. Veri kümesindeki toplam yinelenen satır sayısını sayalım
duplicate_count = df.duplicated().sum()
print(f"Veri kümesindeki toplam yinelenen satır sayısı: {duplicate_count}")

# 2. Eğer varsa, bu yinelenen satırlara bir göz atalım
if duplicate_count > 0:
    print("\nYinelenen satırlardan ilk 5 örnek:")
    display(df[df.duplicated()].head())
else:
    print("\nYinelenen satır bulunamadı.")

Görev 2: Veri çerçevesinden yinelenen satırları kaldırın.

```python
# 1. Yinelenen satırları kaldıralım
df.drop_duplicates(inplace=True)

# 2. Kaldırma işleminden sonra tekrar kontrol ederek işlemi doğrulayalım
final_duplicates = df.duplicated().sum()
print(f"Kaldırma işlemi sonrası kalan yinelenen satır sayısı: {final_duplicates}")
print(f"Veri setinin yeni boyutu (Satır, Sütun): {df.shape}")
```

## Adım 4: Eksik Değerleri Bulma
Görev 3: Tüm sütunlardaki eksik değerleri bulun.

```python
# 1. Tüm sütunlar için eksik değerlerin (null) toplam sayısını belirleyelim
missing_values_all = df.isnull().sum()

# 2. Sonuçları daha iyi analiz etmek için sadece eksik değeri olan ilk 20 sütunu görelim
print("Sütun bazında eksik değerlerin sayıları (İlk 20):")
print(missing_values_all[missing_values_all > 0].head(20))

# 3. Veri setindeki toplam eksik hücre sayısını da görelim
print(f"\nVeri setindeki toplam toplam eksik hücre sayısı: {missing_values_all.sum()}")
```


Görev 4: RemoteWork sütununda kaç satırın eksik olduğunu bulun.

```python
# 1. RemoteWork sütunundaki eksik değerleri sayalım
remote_work_missing = df['RemoteWork'].isnull().sum()

print(f"'RemoteWork' sütunundaki toplam eksik satır sayısı: {remote_work_missing}")

# 2. Ek bilgi: Bu eksiğin tüm veri setine oranını hesaplayalım
remote_work_ratio = (remote_work_missing / len(df)) * 100
print(f"RemoteWork sütunundaki eksiklik oranı: %{remote_work_ratio:.2f}")
```

## Adım 5. Eksik Değerlerin Doldurulması
Görev 5: RemoteWork sütunundaki değer sayılarını bulun.

```python
# Her bir kategorinin kaç kez tekrar ettiğini bulalım
remote_work_counts = df['RemoteWork'].value_counts()
print("RemoteWork Kategorileri ve Kişi Sayıları:")
print(remote_work_counts)
```

Görev 6: Uzaktan Çalışma sütununda en sık (çoğunlukla) görülen değeri belirleyin.

```python
# En sık tekrar eden değeri (Mod) bulalım
remote_mode = df['RemoteWork'].mode()[0]
print(f"\nRemoteWork sütunundaki en yaygın (çoğunluk) değer: {remote_mode}")
```

Görev 7: RemoteWork sütunundaki tüm boş satırları çoğunluk değeriyle doldurun (değiştirin).

```python
# Eksik değerleri mod (çoğunluk) değeri ile dolduralım
df['RemoteWork'].fillna(remote_mode, inplace=True)

# İşlemi doğrulayalım
print(f"Doldurma sonrası RemoteWork sütununda kalan eksik sayısı: {df['RemoteWork'].isnull().sum()}")
```

Görev 8: Ücretlendirmeyle ilgili sütunları kontrol edin ve dağılımlarını açıklayın.

```python
# İçinde "Comp" veya "Currency" geçen tüm sütunları bulalım
actual_salary_cols = [col for col in df.columns if 'Comp' in col or 'Currency' in col]
print("Veri setindeki gerçek ücretlendirme sütunları:")
print(actual_salary_cols)
```

```python
# 1. Gerçek ücretlendirme sütunlarını seçelim
salary_cols = ['Currency', 'CompTotal', 'ConvertedCompYearly']

# 2. Eksik değerleri ve temel istatistikleri görelim
print("### Ücretlendirme Sütunları Analizi ###")
print(df[salary_cols].isnull().sum())

print("\n### Sayısal Dağılım Özeti ###")
display(df[['CompTotal', 'ConvertedCompYearly']].describe())

# 3. Para birimi çeşitliliğini kontrol edelim (Dağılım açıklaması için önemli)
print("\n### En Çok Kullanılan 5 Para Birimi ###")
print(df['Currency'].value_counts().head(5))
```
