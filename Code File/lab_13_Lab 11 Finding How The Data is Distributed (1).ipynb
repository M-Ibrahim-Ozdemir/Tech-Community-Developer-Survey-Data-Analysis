## Adım 1: Kütüphaneleri İçe Aktarın ve Verileri Yükleyin
Pandas, matplotlib.pyplot ve seaborn kütüphanelerini içe aktarın.
Veri setini yükleyerek başlayacaksınız. JupyterLite üzerinde çalışıyorsanız pyfetch yöntemini kullanabilirsiniz. Aksi takdirde, yerel makinelerinizde veya bulut ortamlarınızda doğrudan pandas'ın read_csv() fonksiyonunu kullanabilirsiniz.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

# Adım 2: Veri Yapısını İnceleyin
-Veri yapısını anlamak için sütun adlarını, veri türlerini ve özet bilgilerini görüntüleyin.

-Amaç: Veri kümesinin şekli ve mevcut değişkenler hakkında bilgi edinmek.

```python
# 1. Veri kümesinin satır ve sütun sayısını görelim
print(f"Veri Kümesi Şekli (Satır, Sütun): {df.shape}")

# 2. Sütun adlarını ve veri türlerini (Dtype) listeleyelim
# Bu, hangi sütunlarla matematiksel işlem yapabileceğimizi söyler
print("\n### Sütun Bilgileri ve Veri Türleri ###")
df.info()

# 3. Sayısal sütunlar için hızlı bir özet istatistik (count, mean, std, min, max)
# Dağılım hakkında ilk ipucunu buradan alırız
print("\n### Sayısal Sütunların Özet İstatistikleri ###")
display(df.describe())
```

# 3. Adım: Eksik Verilerin Ele Alınması
-Veri kümesindeki eksik değerleri belirleyin.

-Veri bütünlüğünü sağlamak için gerektiği gibi eksik değerleri doldurun veya kaldırın.

```python
# 1. Eksik verilerin yüzdesini kontrol edelim (Hangi sütunlar çok boş?)
missing_data = df.isnull().mean() * 100
cols_to_drop = missing_data[missing_data > 70].index # %70'ten fazlası boş olanları seç

print(f"Silinecek Sütun Sayısı (%70+ boş olanlar): {len(cols_to_drop)}")

# 2. Çok boş olan sütunları veri setinden atalım
df.drop(columns=cols_to_drop, inplace=True)

# 3. Kritik Sayısal Sütunları Medyan ile dolduralım
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

# 4. Kritik Kategorik Sütunları Mod ile dolduralım
categorical_cols = ['Country', 'EdLevel', 'Employment', 'RemoteWork', 'JobSat']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])

# 5. Sonuç Kontrolü
print("\nKritik sütunlarda kalan eksik veri sayısı:")
print(df[categorical_cols].isnull().sum())
print("\nToplam Sayısal Eksik Veri:", df[numeric_cols].isnull().sum().sum())
```

# 5. Adım: İş Memnuniyetini Görselleştirme (JobSat'a Odaklanın)
JobSat'ın dağılımını görselleştirmek için bir pasta grafiği veya KDE grafiği oluşturun.

Grafiğin yorumunu yapın ve iş memnuniyetindeki temel eğilimleri vurgulayın.


```python
# 1. KDE Grafiği (Yoğunluk Dağılımı)
plt.figure(figsize=(12, 6))
sns.kdeplot(df['JobSat'], fill=True, color="dodgerblue", bw_adjust=0.5)   #fill içinidolduruyor

# Ortalama ve Medyan çizgilerini ekleyerek farkı görelim
plt.axvline(df['JobSat'].mean(), color='red', linestyle='--', label=f'Ortalama: {df["JobSat"].mean():.2f}')
plt.axvline(df['JobSat'].median(), color='green', linestyle='-', label=f'Medyan: {df["JobSat"].median():.2f}')

plt.title('İş Memnuniyeti Puanlarının Dağılım Yoğunluğu (KDE)', fontsize=15)
plt.xlabel('İş Memnuniyeti Puanı (0-10)', fontsize=12)
plt.ylabel('Yoğunluk', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()

# 2. Pasta Grafiği İçin Gruplandırma (Özet Görünüm)
# Puanları Düşük (0-4), Orta (5-7) ve Yüksek (8-10) olarak gruplayalım
bins = [-1, 4, 7, 10]
labels = ['Düşük (0-4)', 'Orta (5-7)', 'Yüksek (8-10)']
df['JobSat_Group'] = pd.cut(df['JobSat'], bins=bins, labels=labels)

plt.figure(figsize=(8, 8))
df['JobSat_Group'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightcoral', 'skyblue', 'lightgreen'], startangle=140)
plt.title('İş Memnuniyeti Gruplarına Göre Dağılım', fontsize=15)
plt.ylabel('') # Y ekseni etiketini kaldıralım
plt.show()
```

## Adım 6: Programlama Dilleri Analizi
LanguageHaveWorkedWith ve LanguageWantToWorkWith tablolarındaki programlama dillerinin sıklığını karşılaştırın.

Örtüşmeleri veya farklılıkları Venn diyagramı veya gruplandırılmış çubuk grafik kullanarak görselleştirin.

```python
# 1. Dilleri ayırıp tek tek sayan bir fonksiyon tanımlayalım
def get_individual_languages(column_name):
    # Hücre içindeki 'Python;Java' gibi yapıları ayırıp her dili tek satıra indirir
    return df[column_name].str.split(';').explode().value_counts()

# 2. İki sütun için de dil sayılarını hesaplayalım
have_worked = get_individual_languages('LanguageHaveWorkedWith')
want_to_work = get_individual_languages('LanguageWantToWorkWith')
print(f"Have worked(su an kullanılan) : {have_worked}")
print("*************************************************************************")
print(f"want to work (Gelecekte istenen): {want_to_work} ")


# 3. Bu iki veriyi tek bir tabloda birleştirelim
lang_df = pd.DataFrame({
    'HaveWorkedWith': have_worked,
    'WantToWorkWith': want_to_work
}).fillna(0) # Hiç seçilmeyen dillere 0 yazalım

# 4. En popüler 15 dili seçelim (Grafiğin okunabilir olması için)
top_15_langs = lang_df.sort_values(by='HaveWorkedWith', ascending=False).head(15)

# 5. Görselleştirme: Gruplandırılmış Çubuk Grafik
plt.figure(figsize=(12, 8))
top_15_langs.plot(kind='barh', width=0.8, color=['#3498db', '#e74c3c'], ax=plt.gca())

plt.title('Mevcut Kullanılan Diller vs. Öğrenilmek İstenen Diller (Top 15)', fontsize=15)
plt.xlabel('Kişi Sayısı', fontsize=12)
plt.ylabel('Programlama Dilleri', fontsize=12)
plt.legend(['Şu An Kullanılan', 'Gelecekte İstenen'])
plt.gca().invert_yaxis() # En popüler dili en üste alalım
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

 ## Adım 7: Uzaktan Çalışma Trendlerini Analiz Edin
Gruplandırılmış çubuk grafik veya ısı haritası kullanarak bölgelere göre uzaktan çalışma dağılımını görselleştirin.

```python
#Bu kod, en çok katılım sağlayan ilk 10 ülkeyi seçer ve bu ülkelerdeki çalışma modellerinin dağılımını yüzde (%) olarak görselleştirir:
# 1. En çok katılımcısı olan ilk 10 ülkeyi belirleyelim
top_10_countries = df['Country'].value_counts().index[:10]
subset_df = df[df['Country'].isin(top_10_countries)]   #yani ulke sutunu filtreledim isin yaptım sadece şleme bu ilk 10u alcak

# 2. Ülke ve Uzaktan Çalışma modeli bazında bir Çapraz Tablo (Crosstab) oluşturalım
# normalize='index' kullanarak sayıları yüzdeye çeviriyoruz (Ülkelerin nüfus farkı analizi bozmasın diye)
remote_ct = pd.crosstab(subset_df['Country'], subset_df['RemoteWork'], normalize='index') * 100

# 3. Görselleştirme: Isı Haritası (Heatmap)
plt.figure(figsize=(12, 8))
sns.heatmap(remote_ct, annot=True, fmt='.1f', cmap='YlGnBu', cbar_kws={'label': 'Yüzde (%)'})

plt.title('Ülkelere Göre Uzaktan Çalışma Modellerinin Dağılımı (%)', fontsize=16)
plt.xlabel('Çalışma Modeli', fontsize=12)
plt.ylabel('Ülke', fontsize=12)
plt.show()
```

## Adım 8: İş Memnuniyeti ve Deneyim Arasındaki Korelasyon
Genel iş memnuniyeti (JobSat) ile YearsCodePro arasındaki korelasyonu analiz edin.

Pearson veya Spearman korelasyon katsayısını hesaplayın.

Pearson korelasyonu doğrusal ilişkileri ölçerken, Spearman verideki sıralamayı (eğilimi) baz alır. Genellikle anket verilerinde Spearman daha sağlam sonuçlar verebilir ancak biz her ikisini de görelim.

```python
!pip install scipy
```

```python
# 1. Verileri sayısal formata hazırlayalım (Errors='coerce' hatalı verileri NaN yapar)
df['YearsCodePro_Numeric'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
df['JobSat_Numeric'] = pd.to_numeric(df['JobSat'], errors='coerce')

# 2. Korelasyon katsayılarını hesaplayalım (Boş değerleri atlayarak)
pearson_corr = df['JobSat_Numeric'].corr(df['YearsCodePro_Numeric'], method='pearson')
spearman_corr = df['JobSat_Numeric'].corr(df['YearsCodePro_Numeric'], method='spearman')

print(f"Pearson Korelasyon Katsayısı: {pearson_corr:.4f}")
print(f"Spearman Korelasyon Katsayısı: {spearman_corr:.4f}")

# 3. Görselleştirme: Regresyon Çizgili Dağılım Grafiği (Joint Plot)
# Bu grafik hem korelasyonu hem de her iki değişkenin dağılımını aynı anda gösterir
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='YearsCodePro_Numeric', y='JobSat_Numeric', 
            scatter_kws={'alpha':0.1, 's':10}, line_kws={'color':'red'})

plt.title('Deneyim Yılları ve İş Memnuniyeti Arasındaki İlişki', fontsize=14)
plt.xlabel('Profesyonel Deneyim (Yıl)', fontsize=12)
plt.ylabel('İş Memnuniyeti Puanı', fontsize=12)
plt.show()
```

## Adım 9: Çapraz Tablo Analizi (İstihdam ve Eğitim Seviyesi)
İstihdam durumu (İstihdam) ve eğitim seviyesi (Eğitim Seviyesi) arasındaki ilişkiyi analiz edin.

Talimat: pd.crosstab() kullanarak bir çapraz tablo oluşturun ve mümkünse yığılmış çubuk grafik ile görselleştirin.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. EDA Hazırlığı: Uzun metinleri gruplayıp kısaltalım
# Eğitim seviyelerini kısaltmak için bir fonksiyon
def shorten_ed(text):
    if pd.isna(text): return "Other"
    if "Bachelor" in text: return "Bachelor's"
    if "Master" in text: return "Master's"
    if "Professional degree" in text: return "PhD/Prof"
    if "Secondary school" in text: return "High School"
    if "Some college" in text: return "Some College"
    return "Other"

# İstihdam durumlarını kısaltmak için bir fonksiyon
def shorten_emp(text):
    if pd.isna(text): return "Other"
    first = text.split(';')[0]
    if "full-time" in first: return "Full-time"
    if "Independent" in first: return "Freelancer"
    if "part-time" in first: return "Part-time"
    if "Student" in first: return "Student"
    return "Other"

df_eda = df.copy()
df_eda['Ed_Clean'] = df_eda['EdLevel'].apply(shorten_ed)
df_eda['Emp_Clean'] = df_eda['Employment'].apply(shorten_emp)

# 2. Çapraz Tablo (Crosstab) Oluşturma
# normalize='index' yaparak her eğitim seviyesini kendi içinde %100 üzerinden değerlendirelim
eda_crosstab = pd.crosstab(df_eda['Ed_Clean'], df_eda['Emp_Clean'], normalize='index') * 100

# 3. Görselleştirme (Yığılmış Çubuk Grafik)


ax = eda_crosstab.plot(kind='barh', stacked=True, figsize=(12, 7), 
                        color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])

plt.title('Eğitim Seviyesi ve İstihdam Durumu İlişkisi (%)', fontsize=14, fontweight='bold')
plt.xlabel('Yüzde (%)')
plt.ylabel('Eğitim Seviyesi')
plt.legend(title='İstihdam Türü', bbox_to_anchor=(1.05, 1), loc='upper left')

# Yüzde değerlerini sadece görünür alanlara ekle
for p in ax.patches:
    width = p.get_width()
    if width > 5: # Sadece %5'ten büyük olanlara yaz
        ax.annotate(f'{int(width)}%', 
                    (p.get_x() + width / 2., p.get_y() + p.get_height() / 2.), 
                    ha='center', va='center', color='white', fontweight='bold')

plt.tight_layout()
plt.show()
```
