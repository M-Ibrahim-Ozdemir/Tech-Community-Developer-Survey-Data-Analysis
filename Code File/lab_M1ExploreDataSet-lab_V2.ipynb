Anket katılımcılarının ortalama yaşını yazdırın.

```python
# print("Anket katılımcılarının ortalama yaşı:", df['Age'].mean())  bu olmaz cunku ıcnde string varmıs mecbur dnosturdecz sonra eskihaline almalıyız

import pandas as pd

# 1. Veriyi en baştan, temiz haliyle tekrar okuyalım (NaN hatasını düzeltmek için şart)
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv"
df = pd.read_csv(file_path)

# 2. Yaş aralıklarını temsil eden sayısal orta noktaları tanımlayalım
# Bu orta noktalar anket analizlerinde standart kabul edilir.
age_map = {
    'Under 18 years old': 15,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70
}

# 3. Metinleri sayılara haritalayalım
df['Age_Numeric'] = df['Age'].map(age_map)

# 4. Sayısal ortalamayı hesaplayalım
ortalama = df['Age_Numeric'].mean()

print(f"Hesaplanan Yaklaşık Ortalama Yaş: {ortalama:.1f}")
```

```python
#hesapladıgımız veriyisilmeliyiz
# Mükerrer satır sayısını bul
duplicate_count = df.duplicated().sum()
print("Veri setindeki mükerrer satır sayısı:", duplicate_count)
```

```python
# Mükerrer satırları sil ve yeni bir dataframe'e kaydet
df_cleaned = df.drop_duplicates()

# Kontrol etmek için yeni boyutu yazdır
print("Temizlik sonrası veri seti boyutu:", df_cleaned.shape)
```
