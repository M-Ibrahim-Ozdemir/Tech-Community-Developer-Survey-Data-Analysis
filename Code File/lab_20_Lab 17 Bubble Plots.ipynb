## Görev 1: Baloncuk Grafikleri Kullanarak Veri Dağılımlarını İnceleme
***1. Yaş ve Katılım Sıklığı Arasındaki İlişkinin Baloncuk Grafiği***

Katılımcıların yaşı ile katılım sıklığı (SOPartFreq) arasındaki ilişkiyi bir baloncuk grafiği kullanarak görselleştirin.

Baloncukların boyutunu iş memnuniyetlerini (JobSat) temsil etmek için kullanın.

```python
#Bu grafik, "Yaşlandıkça topluluğa katılım azalıyor mu ve bu mutluluğu nasıl etkiliyor?" sorusuna yanıt verir.
print(df['SOPartFreq'].value_counts())
print('*******************************************')
print(df['JobSat'].value_counts())
print('*******************************************')
print(df['Age'].value_counts())

print('********----------************-------------**********---------------************')

# 1. Yaş gruplarını mantıksal sıraya koyalım
age_order = [
    'Under 18 years old', '18-24 years old', '25-34 years old', 
    '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older'
]

# 2. Katılım Sıklığını (SOPartFreq) sayısal bir sıraya dizelim
# En azdan en çoğa doğru bir puanlama yapıyoruz
freq_map = {
    'I have never participated in Q&A on Stack Overflow': 0,
    'Less than once per month or monthly': 1,
    'A few times per month or weekly': 2,
    'A few times per week': 3,
    'Daily or almost daily': 4,
    'Multiple times per day': 5
}
df['SOPartFreq_Score'] = df['SOPartFreq'].map(freq_map)

# 3. Veriyi temizleyelim (Boş olanları atalım)
df_bubble = df.dropna(subset=['Age', 'SOPartFreq_Score', 'JobSat']).copy()

# Yaş sütununu sıralı kategoriye çevirelim
df_bubble['Age'] = pd.Categorical(df_bubble['Age'], categories=age_order, ordered=True)

# 4. Görselleştirme
plt.figure(figsize=(14, 8))

# s: Kabarcık boyutu (İş memnuniyeti puanını belirgin olması için 50 ile çarpıyoruz)
# hue: Renk (Yine memnuniyet puanına göre renk değişsin)
sns.scatterplot(data=df_bubble, x='Age', y='SOPartFreq_Score', 
                size='JobSat', hue='JobSat',
                sizes=(20, 400), alpha=0.5, palette='viridis')

plt.title('Yaş ve Katılım Sıklığı Arasındaki İlişki (Boyut: İş Memnuniyeti)', fontsize=14)
plt.xlabel('Yaş Grubu')
plt.ylabel('Katılım Sıklığı Puanı (0: Hiç - 5: Çok)')
plt.xticks(rotation=30)
plt.grid(axis='both', alpha=0.3)

# Lejantı düzenleyelim
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='İş Memnuniyeti (0-10)')
plt.tight_layout()
plt.show()
```

Eğer bir sütundaki balonlar genellikle daha büyükse, o yaş grubunun genel olarak işinden daha memnun olduğunu kolayca anlayabiliriz

***2. Ücret ve İş Memnuniyeti Arasındaki İlişkiyi Gösteren Balon Grafiği***
- Yıllık ücret (ConvertedCompYearly) ve iş memnuniyeti (JobSat) arasındaki ilişkiyi görselleştirin.

Balonların boyutunu katılımcıların yaşını temsil etmek için kullanın.

```python
print(df['ConvertedCompYearly'].value_counts())
print('*******************************************')
print(df['JobSat'].value_counts())
print('*******************************************')
print(df['Age'].value_counts())
print('*******************************************')

# 1. Yaş gruplarının mantıksal sırasını tanımlayalım
age_order = [
 
    '   'Under 18 years old', 18-24 years old', 
    '25-34 years old', 
    '35-44 years old', 
    '45-54 years old', 
    '55-64 years old', 
    '65 years or older'
]

# 2. Boyut eşlemesini (Age_Size) tekrar tanımlayalım
age_size_map = {
    'Under 18 years old': 50,
    '18-24 years old': 120,
    '25-34 years old': 250,
    '35-44 years old': 450,
    '45-54 years old': 700,
    '55-64 years old': 1000,
    '65 years or older': 1400
}

# 3. Veriyi hazırlama ve SIRALAMA
# Yaş sütununu kategorik ve sıralı yapıyoruz (KRİTİK ADIM)
df['Age'] = pd.Categorical(df['Age'], categories=age_order, ordered=True)
df['Age_Size'] = df['Age'].map(age_size_map)

# Temizlik ve filtreleme
df_clean = df.dropna(subset=['ConvertedCompYearly', 'JobSat', 'Age_Size']).copy()
upper_limit = df_clean['ConvertedCompYearly'].quantile(0.95)
df_filtered = df_clean[df_clean['ConvertedCompYearly'] <= upper_limit].sort_values('Age')

# 4. Görselleştirme
plt.figure(figsize=(16, 10))

# hue_order ekleyerek renklerin de tam istediğimiz sırada olmasını garanti ediyoruz
scatter = sns.scatterplot(data=df_filtered, 
                          x='ConvertedCompYearly', 
                          y='JobSat', 
                          size='Age_Size', 
                          hue='Age',
                          hue_order=age_order, # Renk sırasını sabitledik
                          sizes=(50, 1100), 
                          alpha=0.4, 
                          palette='viridis')

plt.title('Yıllık Ücret vs İş Memnuniyeti (Sıralı Yaş Grupları)', fontsize=16)
plt.xlabel('Yıllık Ücret (USD)', fontsize=12)
plt.ylabel('İş Memnuniyeti (0 - 10 Puan)', fontsize=12)

# X ekseni formatı (100,000 gibi)
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))

# Lejant düzenleme: Sadece yaş gruplarını gösterelim, boyut etiketlerini gizleyelim
handles, labels = scatter.get_legend_handles_labels()
# Sadece ilk 7 öğeyi (yaş grupları) alıyoruz, boyut karmaşasını atıyoruz
plt.legend(handles[1:8], labels[1:8], bbox_to_anchor=(1.02, 1), loc='upper left', title='Yaş Grupları')

plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

# Görev 2: Balon Grafikleri Kullanarak İlişkileri Analiz Etme
***1. Yaşa Göre Teknoloji Tercihlerinin Balon Grafiği***

Katılımcıların çalıştığı programlama dillerinin (LanguageHaveWorkedWith) yaş gruplarına göre popülerliğini görselleştirin.

Her dilin sıklığını temsil etmek için balon boyutunu kullanın.

```python
print(df['LanguageHaveWorkedWith'].value_counts())
print('*******************************************')

print(df['DatabaseWantToWorkWith'].value_counts())
print('*******************************************')

# 1. Veriyi hazırla ve dilleri ayır
df_lang = df.dropna(subset=['LanguageHaveWorkedWith', 'Age']).copy()
df_lang['Language'] = df_lang['LanguageHaveWorkedWith'].str.split(';')
df_lang_exploded = df_lang.explode('Language')

# 2. En popüler 10 dili seçelim (Grafik çok karışık olmasın diye)
top_10_langs = df_lang_exploded['Language'].value_counts().nlargest(10).index
df_lang_top = df_lang_exploded[df_lang_exploded['Language'].isin(top_10_langs)].copy()

# 3. Yaş ve Dil bazında sayıları hesapla (Balon boyutu için)
df_counts = df_lang_top.groupby(['Age', 'Language']).size().reset_index(name='Count')

# Yaş sıralamasını 
age_order = ['Under 18 years old', '18-24 years old', '25-34 years old', '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older']
df_counts['Age'] = pd.Categorical(df_counts['Age'], categories=age_order, ordered=True)

# 4. Görselleştirme
plt.figure(figsize=(16, 8))
scatter = sns.scatterplot(data=df_counts, x='Language', y='Age', 
                          size='Count', hue='Language',
                          sizes=(100, 2000), alpha=0.6, palette='tab10')

plt.title('Yaş Gruplarına Göre Programlama Dili Popülerliği', fontsize=15)
plt.xlabel('Programlama Dili')
plt.ylabel('Yaş Grubu')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', title='Kullanım Sayısı')
plt.grid(axis='both', alpha=0.2)
plt.show()
```

***2. Tercih Edilen Veritabanları(DatabaseWantToWorkWith) ve İş Memnuniyeti Arasındaki İlişkiyi Gösteren Balon Grafiği***

Tercih edilen veritabanları (Çalışmak İstediğim Veritabanı) ve iş memnuniyeti arasındaki ilişkiyi inceleyin.

Her veritabanı için yanıt verenlerin sayısını belirtmek için balon boyutunu kullanın.


```python
# 1. Veriyi hazırla ve veritabanlarını ayır
df_db = df.dropna(subset=['DatabaseWantToWorkWith', 'JobSat']).copy()
df_db['Database'] = df_db['DatabaseWantToWorkWith'].str.split(';')
df_db_exploded = df_db.explode('Database')

# 2. En popüler 10 veritabanını seçelim
top_10_dbs = df_db_exploded['Database'].value_counts().nlargest(10).index
df_db_top = df_db_exploded[df_db_exploded['Database'].isin(top_10_dbs)]

# 3. Her veritabanı için ortalama memnuniyet ve kişi sayısını hesapla
db_stats = df_db_top.groupby('Database').agg(
    Avg_JobSat=('JobSat', 'mean'),
    User_Count=('Database', 'count')
).reset_index()

# 4. Görselleştirme
plt.figure(figsize=(14, 7))
bubble = sns.scatterplot(data=db_stats, x='Database', y='Avg_JobSat', 
                         size='User_Count', hue='Database',
                         sizes=(200, 3000), alpha=0.6, palette='viridis', legend=False)

plt.title('Tercih Edilen Veritabanları ve Ortalama İş Memnuniyeti', fontsize=15)
plt.xlabel('Veritabanı')
plt.ylabel('Ortalama İş Memnuniyeti (0-10)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Her balonun üzerine veritabanı ismini yazalım 
for i in range(db_stats.shape[0]):
    plt.text(db_stats.Database[i], db_stats.Avg_JobSat[i], db_stats.Database[i], 
             ha='center', va='center', fontsize=9, fontweight='bold')

plt.show()
```

# Görev 3: Balon Grafikleri Kullanarak Verileri Karşılaştırma
***1. Geliştirici Rollerine Göre Ücretlendirme için Balon Grafiği***

Farklı geliştirici rollerine (DevType) göre ücretlendirmeyi (ConvertedCompYearly) görselleştirin.

İş memnuniyetini temsil etmek için balon boyutunu kullanın. job satisfaction.

```python
print(df['DevType'].value_counts())
print('*******************************************')
print(df['ConvertedCompYearly'].value_counts())
print('*******************************************')
print(df['JobSat'].value_counts())
print('*******************************************')
# 1. Veriyi temizle ve en popüler 10 rolü seç
df_dev = df.dropna(subset=['DevType', 'ConvertedCompYearly', 'JobSat']).copy()
top_10_roles = df_dev['DevType'].value_counts().nlargest(10).index
df_dev_top = df_dev[df_dev['DevType'].isin(top_10_roles)].copy()

# 2. Maaş uç değerlerini temizle (%95 dilimi)
limit = df_dev_top['ConvertedCompYearly'].quantile(0.95)
df_dev_filtered = df_dev_top[df_dev_top['ConvertedCompYearly'] <= limit]

# 3. Görselleştirme
plt.figure(figsize=(16, 9))

# s: JobSat (boyut), hue: DevType (renk)
sns.scatterplot(data=df_dev_filtered, x='DevType', y='ConvertedCompYearly', 
                size='JobSat', hue='DevType',
                sizes=(10, 1500), alpha=0.5, palette='tab10')

plt.title('Geliştirici Rollerine Göre Maaş Dağılımı (Boyut: İş Memnuniyeti)', fontsize=15)
plt.xlabel('Geliştirici Rolü')
plt.ylabel('Yıllık Ücret (USD)')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', title='Roller / Memnuniyet')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

***2. Yaş Gruplarına Göre İşbirliği Araçları için Balon Grafiği***

Kullanılan işbirliği araçları (NEWCollabToolsHaveWorkedWith) ile yaş grupları arasındaki ilişkiyi görselleştirin.

Araç kullanım sıklığını temsil etmek için balon boyutunu kullanın.

```python
print(df['NEWCollabToolsHaveWorkedWith'].value_counts())
print('*******************************************')

# 1. Veriyi hazırla ve araçları ayır
df_tools = df.dropna(subset=['NEWCollabToolsHaveWorkedWith', 'Age']).copy()
df_tools['Tool'] = df_tools['NEWCollabToolsHaveWorkedWith'].str.split(';')
df_tools_exploded = df_tools.explode('Tool')

# 2. En popüler 10 aracı seçelim
top_10_tools = df_tools_exploded['Tool'].value_counts().nlargest(10).index
df_tools_top = df_tools_exploded[df_tools_exploded['Tool'].isin(top_10_tools)].copy()

# 3. Yaş ve Araç bazında kullanım sayılarını hesapla
tool_counts = df_tools_top.groupby(['Age', 'Tool']).size().reset_index(name='Usage_Count')

# Yaş sıralamasını uygula
age_order = ['Under 18 years old', '18-24 years old', '25-34 years old', 
             '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older']
tool_counts['Age'] = pd.Categorical(tool_counts['Age'], categories=age_order, ordered=True)

# 4. Görselleştirme
plt.figure(figsize=(16, 8))
sns.scatterplot(data=tool_counts, x='Tool', y='Age', 
                size='Usage_Count', hue='Tool',
                sizes=(100, 2500), alpha=0.6, palette='Set2')

plt.title('Yaş Gruplarına Göre İşbirliği Araçları Kullanımı', fontsize=15)
plt.xlabel('İşbirliği Aracı (IDE/Editor)')
plt.ylabel('Yaş Grubu')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', title='Kullanım Sayısı')
plt.grid(axis='both', alpha=0.2)
plt.tight_layout()
plt.show()
```

# Görev 4: Balon Grafikleri Kullanarak Teknoloji Trendlerini Görselleştirme
***1. Tercih Edilen Web Çerçeveleri ve İş Memnuniyeti Arasındaki İlişkinin Balon Grafiği***

Tercih edilen web çerçeveleri (WebframeWantToWorkWith) ile iş memnuniyeti arasındaki ilişkiyi inceleyin.

Katılımcı sayısını temsil etmek için balon boyutunu kullanın.


```python
print(df['WebframeWantToWorkWith'].value_counts())
print('*******************************************')

# 1. Veriyi hazırla ve Web Çerçevelerini ayır
df_web = df.dropna(subset=['WebframeWantToWorkWith', 'JobSat']).copy()
df_web['Webframe'] = df_web['WebframeWantToWorkWith'].str.split(';')
df_web_exploded = df_web.explode('Webframe')

# 2. En popüler 10 çerçeveyi seçelim
top_10_web = df_web_exploded['Webframe'].value_counts().nlargest(10).index
df_web_top = df_web_exploded[df_web_exploded['Webframe'].isin(top_10_web)]

# 3. İstatistikleri hesapla
web_stats = df_web_top.groupby('Webframe').agg(
    Avg_JobSat=('JobSat', 'mean'),
    Count=('Webframe', 'count')
).reset_index()

# 4. Görselleştirme
plt.figure(figsize=(14, 7))
sns.scatterplot(data=web_stats, x='Webframe', y='Avg_JobSat', 
                size='Count', hue='Webframe',
                sizes=(40, 3500), alpha=0.6, palette='Set1', legend=False)

plt.title('Gelecekte Tercih Edilen Web Çerçeveleri ve İş Memnuniyeti', fontsize=15)
plt.ylabel('Ortalama İş Memnuniyeti (0-10)')
plt.xticks(rotation=30)

# İsimleri balonların üzerine yazalım
for i in range(web_stats.shape[0]):
    plt.text(web_stats.Webframe[i], web_stats.Avg_JobSat[i], web_stats.Webframe[i], 
             ha='center', va='center', fontsize=10, fontweight='bold')

plt.show()
```

***2. Ülkeler Arasında Beğenilen Teknolojiler İçin Balon Grafiği***

Beğenilen teknolojilerin (LanguageAdmired) farklı ülkeler (Country) genelindeki dağılımını görselleştirin.

Beğenilme sıklığını temsil etmek için balon boyutunu kullanın.

```python
print(df['LanguageAdmired'].value_counts())
print('*******************************************')
print(df['Country'].value_counts())
print('*******************************************')

# 1. Veriyi hazırla ve Beğenilen Dilleri ayır
df_admired = df.dropna(subset=['LanguageAdmired', 'Country']).copy()
df_admired['Language'] = df_admired['LanguageAdmired'].str.split(';')
df_admired_exploded = df_admired.explode('Language')

# 2. En popüler 5 ülke ve en popüler 10 dili filtreleyelim (Karmaşayı önlemek için)
top_countries = df_admired_exploded['Country'].value_counts().nlargest(5).index
top_languages = df_admired_exploded['Language'].value_counts().nlargest(10).index

df_final = df_admired_exploded[
    (df_admired_exploded['Country'].isin(top_countries)) & 
    (df_admired_exploded['Language'].isin(top_languages))
]

# 3. Sayıları hesapla
admired_counts = df_final.groupby(['Country', 'Language']).size().reset_index(name='Freq')

# 4. Görselleştirme
plt.figure(figsize=(16, 8))
sns.scatterplot(data=admired_counts, x='Language', y='Country', 
                size='Freq', hue='Country',
                sizes=(100, 3000), alpha=0.6, palette='tab10')

plt.title('Ülkeler Arasında Beğenilen Yazılım Dilleri Dağılımı', fontsize=15)
plt.xlabel('Beğenilen Dil')
plt.ylabel('Ülke')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', title='Beğenilme Sıklığı')
plt.grid(axis='both', alpha=0.2)
plt.tight_layout()
plt.show()

```
